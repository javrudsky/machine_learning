```python
# Getting the mnist data
from sklearn.datasets import fetch_openml
import numpy as np
import pandas as pd

# first try to load data from disk
X = None
y = None
mnist = None
try:
    X = pd.read_csv('mnist_data.csv')
except FileNotFoundError:
    print("Mnist data not available")

if X is None:
    # Fetch from remote
    mnist = fetch_openml('mnist_784', parser='auto', version=1)
    # Splitting into train and test data.
    # MNIST dataset has already done, so just slicing it
    X, y = mnist["data"], mnist["target"]
    # Saving for next time
    mnist.data.to_csv("mnist_data.csv")
    mnist.target.to_csv("mnist_y.csv")
else:
    # Load labels
    y = pd.read_csv('mnist_y.csv')
    # Fixing some issues with loaded data
    # Removing undesirable rows and converting "y" to a Series object
    X = X.drop(['Unnamed: 0'], axis=1)
    y = y.squeeze() # Converting to a series

# Showing keys
X.keys
```

```python
# This dataset is already split in a training set and test set
# so, let's get them in different variables
# It is important to remark that the dataset is shuffled this way
# so the split is straightforward
train_count=60000
X_train, X_test, y_train, y_test = X[:train_count], X[train_count:], y[:train_count], y[train_count:]
```

```python
# Exercise 2
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.ndimage import shift

def show_digit(digit, reshape=False):
    if reshape:
        digit = digit.reshape(28, 28)
    plt.imshow(digit, cmap=mpl.cm.binary, interpolation="nearest")
    plt.axis("off")
    plt.show()
    
# Now creating the function to shift the digit
# So that we can do data augmentation (or data expansion)
# by adding this new values to our dataset
# direction => [1: right, 2: left, 3: up, 4: down]
def shift_digit(digit, direction=1):
    
    dir_axis = None
    if direction == 1:
        dir_axis = [1, 0]
    elif direction == 2:
        dir_axis = [-1, 0]
    elif direction == 3:
        dir_axis = [0, -1]
    elif direction == 4:
        dir_axis = [0, 1]
    else:
        print("Invalid direction, no shifting done")
        return digit
    
    return shift(digit, shift=dir_axis, mode='constant', cval=0)
    
# Testing the function
digit = X.values[0].reshape(28, 28)
digit = shift_digit(digit, 4)
show_digit(digit)
    
```

```python
# The function was created and it's working.
# Now let's make work on the "dataset expansion"

def expand_dataset(dataset):
    shifted = []
    for digit in dataset.values:
        for direction in range(1,5):
            shifted_d = shift_digit(digit.reshape(28, 28), direction)
            shifted_d = shifted_d.reshape(784)
            shifted.append(shifted_d)
    
    new_ds = dataset.copy()
    ser = pd.DataFrame(data=shifted, columns=new_ds.columns)
    new_ds = pd.concat([new_ds, ser])
    return new_ds   
```

```python
# Loading csv for dataframe

X_exp = None
try:
    X_exp = pd.read_csv('mnist_expanded_dataset.csv', index_col=0)
except FileNotFoundError:
    print("Expanded dataset not available")
    
# If expanded dataset wasn't recreated
# let's create and save it
if X_exp is None:
    # This can take a long time
    X_exp = expand_dataset(X_train)
    # Save the new dataframe
    # Line above is also commented to avoid
    # overwrite generated data
    X_exp.to_csv("mnist_expanded_dataset.csv")
    
X_exp.shape, X_exp.head()
```

```python
# Now we need to expand our labels array
# To avoid a step we also go to convert to numeric our string labels
# y_exp = y.copy()
def expand_labels(labels):
    y_exp = labels.copy()
    for i in range(0, 4):
        y_exp = pd.concat([y_exp, labels])
    return y_exp

y_exp = expand_labels(y_train)
y_exp = y_exp.astype(np.uint8)

# Checking the shape again
y_exp.shape
# Great!
```

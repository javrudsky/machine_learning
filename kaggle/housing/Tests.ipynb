```python
import numpy as np
import matplotlib.pyplot as plt

rad = np.zeros((100, 100))
d = 100
for p in range(100):
    for i in range(100):
        rad[p, i] = p / (np.power(10000, (2 * i / d)))

#print(rad)
#im = plt.pcolormesh(rad, cmap="RdBu")
fig, axes = plt.subplots(1, 3, figsize=(15, 3))
x,y = np.indices(rad.shape)
axes[0].scatter(x,y, s=rad * 5, label="rad")
axes[0].set_xlabel("pos")
axes[0].set_ylabel("index")

#axes[1].plot(rad[::2], np.sin(rad[::2]), label="rad")

print(rad)
ev = rad[2:10:2]
axes[1].plot(ev, np.sin(ev), label="rad")
axes[1].set_xlabel("rad")
axes[1].set_ylabel("sin")

axes[2].plot(rad[1::2], np.cos(rad[1::2]), label="rad")
axes[2].set_xlabel("rad")
axes[2].set_ylabel("cos")
plt.show()
```

```python
# Q-Q plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import stats

n = 100
ds = 0.2
um = 170

rank = np.arange(1, n + 1)
quan = (rank - 0.5) / n
zscore_q = norm.ppf(quan)
zscore = np.round((rank - np.mean(rank)) / np.std(rank), decimals=4) 
vals = np.sort(np.round(np.random.normal(um, ds, n), decimals=2))

nd_slope = (vals[-1] - vals[0]) / (zscore[-1] - zscore[0])
nd_intercept = vals[0] - nd_slope * zscore[0]
nd_vals = nd_slope * zscore + nd_intercept

nd_slope_q = (vals[-1] - vals[0]) / (zscore_q[-1] - zscore_q[0])

nd_intercept_q = vals[0] - nd_slope_q * zscore_q[0]
nd_vals_q = nd_slope_q * zscore_q + nd_intercept_q

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharex=True)

axs[0].scatter(zscore, vals, label="Q-Q plot")
axs[0].plot(zscore, nd_vals, color="green")
axs[0].set_title("Z-score")

axs[1].scatter(zscore_q, vals, label="Q-Q plot (q)")
axs[1].plot(zscore_q, nd_vals_q, color="red")
axs[1].set_title("Z-score Q")

stats.probplot(vals, dist="norm", plot=axs[2])
plt.show()

```
